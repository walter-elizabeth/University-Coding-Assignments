## Question 2: The Bass Diffusion Model (6 points)

One of the earliest applications of modeling to marketing is Frank Bass' 1969 paper studying the spread of new products through a population. The technique is now known as the Bass diffusion model, and it  has been applied to market forecasting, advertising,  economics, and more. The   mathematical model is:

$$A(t) = M\frac{1 - e^{-(p+q)t}}{1 + \frac{q}{p}e^{-(p+q)t}}$$

where $A(t)$ is the expected number of adopters  after time $t$, the **coefficient of innovation**, $p$, is the effect from advertising and "influencers",  the **coefficient of imitation**, $q$, is the effect of  word-of-mouth and followers, and $M$ is the expected market cap, or maximum number of adopters that can be reached.  For this model, we expect the evolution of adopters to behave like so:

```python
#A_t = expected number of adopters after time 𝑡 
#p = coeffiecient of innovation
#q = coeff. of imitation
#M = expected market cap

def bass_model(M, p, q, t):
    A_t = M * ((1 - (math.exp(-(p+q)*t)))/(1 + (q/p)*(math.exp(-(p+q)*t))))
    return A_t

bass_model(1000,0.03,0.38,5)
```

```python
# Part A: Put your pseudocode and code here
#write a function - inputs: a,b. returns 1 or 0. use if/else statement.
import math
def pythag_triple_check(a,b): #function w/inputs a,b
    c = math.sqrt(a**2 + b**2) #define c derived from requirement a^2 + b^2 = c^2
    if c % 1 == 0 and a < b < c:
        return 1
    else:
        return 0
    
pythag_triple_check(3,4)
```

```python
# Part B: Put your pseudocode and code here
#write function- input: n. returns: list of pythagorean triples with c <= n #this will be needed for a loop 
# will have at least 1 loop in a loop
# find c
# return list of all pyth. triple lists within range n
# n is the range of values that c can be

def pythag_triple(n):
    pythag_triples = []
    for a in range(1,n):
        for b in range(a,n):
            for c in range(b+1,n): #need to loop through n
                c = math.sqrt(a * a + b * b) #define c derived from requirement a^2 + b^2 = c^2
            if c % 1 == 0 and a < b < c <= n: # if c is an integer and second part is to confirm that it follows rule of pythag triple and also stops looping when c <= n
                pythag_triples.append([a, b, int(c)]) #append the lists of the triple to the list of lists each time an iteration produces one
                                                        # and int(c) bc sqrt will return a float always and need the integer version; just removes the .0 from c

    print('number of triples:',len(pythag_triples)) # gives no. of triples that are produced in range n
    return pythag_triples
```

## Question 5: The Henon map (7 points)

Suppose you have a point with coordinates $(x_0, y_0)$ on the plane, and that you have specific values for two real numbers $a,b$.
This information is enough to produce two sequences of numbers $x_0, x_1, x_2, \ldots$ and $y_0, y_1, y_2, \ldots$ by using the following equations:

$$
\begin{eqnarray*} 
x_{i+1} &=& 1 \;-\; a\, x_i^2 \;+\; y_i \\ 
y_{i+1} &=& b\, x_i
\end{eqnarray*} 
$$

This kind of iterative process is what people call in mathematics a discrete time __Dynamical System__, where the next value of the $(x_{i+1},y_{i+1})$ coordinates depends on the current value of the $(x_{i},y_{i})$ coordinates. The specific equations shown above are known as the Henon map, and even though it appears to be very simple, the system has a lot of richness controlled by the choices of parameters $a,b$. We will attempt to visualize such complexity in what follows.

```python
def henon(x0, y0, a, b, n):
    x = [x0] #x[i+1] value is alwasy dependent on x[i].... so x[i] always dependent on x[i-1].
    y = [y0]
    x1 = 1 - a * x0**2 + y0 
    y1 = b * x0
    for i in range(n+1):
        x[i] = 1 - a * x[i-1] **2 + y[i-1] #how is it out of range??? smallest i can be is = 1,
        y[i] = b * x[i-1]
        x.append(x[i])
        y.append(y[i])
    return x, y
```
